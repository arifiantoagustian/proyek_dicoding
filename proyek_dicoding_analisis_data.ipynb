## Visualization & Explanatory Analysis

### Pertanyaan 1: Apa produk dengan penjualan terbanyak di e-commerce?

```python
sum_order_items_df = all_data.groupby("product_category_name_english")["product_id"].count().reset_index()
sum_order_items_df = sum_order_items_df.rename(columns={"product_id": "products"})
sum_order_items_df = sum_order_items_df.sort_values(by="products", ascending=False)
sum_order_items_df = sum_order_items_df.head(10)

sum_order_items_df.head()
```

```python
import seaborn as sns

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["#068DA9", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="products", y="product_category_name_english", data=sum_order_items_df.head(5), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Most-Selling Product", loc="center", fontsize=18)
ax[0].tick_params(axis ='y', labelsize=15)

plt.suptitle("Produk dengan Penjualan Tertinggi", fontsize=20)
plt.show()
```

##### Kategori produk dengan penjualan tertinggi di e-commerce adalah pada kategori bed_bath_table dengan total penjualan hampir 4000 produk.

### Pertanyaan 2: Bagaimana kondisi penjualan setiap bulannya?

```python
monthly_df = all_data.resample(rule='M', on='order_approved_at').agg({
    "order_id": "nunique",
})
monthly_df.index = monthly_df.index.strftime('%B') #mengubah format order_approved_at menjadi Tahun-Bulan
monthly_df = monthly_df.reset_index()
monthly_df.rename(columns={
    "order_id": "order_count",
}, inplace=True)
monthly_df.head()
```

```python
monthly_df = monthly_df.sort_values('order_count').drop_duplicates('order_approved_at', keep='last')
```

```python
monthly_df.head()
```

```python
monthly_df.sort_values(by='order_count')
```

```python
month_mapping = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}

monthly_df["month_numeric"] = monthly_df["order_approved_at"].map(month_mapping)
monthly_df = monthly_df.sort_values("month_numeric")
monthly_df = monthly_df.drop("month_numeric", axis=1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_df["order_approved_at"],
    monthly_df["order_count"],
    marker='o',
    linewidth=2,
    color="#068DA9"
)
plt.title("Total Penjualan per Bulan (2018)", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show()
```

##### Total penjualan setiap bulannya cenderung fluktuatif. Penjualan tertinggi terjadi pada bulan Maret dan penjualan terendah terjadi pada bulan September. Berdasarkan diagram garis tersebut, kenaikan penjualan tertinggi terjadi pada blan Oktober-November dan penurunan terbesar terjadi pada bulan Agustus-September.

### Pertanyaan 3: Berapa total uang yang dikeluarkan oleh pelanggan untuk berbelanja di e-commerce dalam 3 bulan terakhir?

```python
monthly_spend_df = all_data.resample(rule='M', on='order_approved_at').agg({
    "payment_value":"sum"
})
monthly_spend_df.index = monthly_spend_df.index.strftime('%B') #mengubah format order_approved_at menjadi Tahun-Bulan
monthly_spend_df = monthly_spend_df.reset_index()
monthly_spend_df.rename(columns={
    "payment_value":"total_spend"
}, inplace=True)
monthly_spend_df.head()
```

```python
monthly_spend_df = monthly_spend_df.sort_values('total_spend').drop_duplicates('order_approved_at', keep='last')
```

```python
monthly_spend_df.head()
```

```python
monthly_spend_df.sort_values(by='total_spend')
```

```python
monthly_spend_df["month_numeric"] = monthly_spend_df["order_approved_at"].map(month_mapping)
monthly_spend_df = monthly_spend_df.sort_values("month_numeric")
monthly_spend_df = monthly_spend_df.drop("month_numeric", axis=1)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_spend_df["order_approved_at"],
    monthly_spend_df["total_spend"],
    marker='o',
    linewidth=2,
    color="#068DA9"
)
plt.title("Total Pengeluaran Uang Pelanggan per Bulan (2018)", loc="center", fontsize=20)
plt.xticks(fontsize=10, rotation=25)
plt.yticks(fontsize=10)
plt.show()
```

##### Dalam tiga bulan terahir, total uang yang dikeluarkan pelanggan untuk berbelanja di e-commerce mengalami kenaikan dan penurunan. Pada bulan Oktober, total uang yang dikeluarkan untuk berbelanja mencapai lebih dari 450000, bulan November mencapai yang tertinggi yaitu lebih dari 500000, dan pada bulan Desember mengalami penurunan dengan total pengeluaran mencapai sekitar 350000.

### Pertanyaan 4: Bagaimana tingkat kepuasan pelanggan terhadap pelayanan yang diberikan e-commerce?

```python
review_scores = all_data['review_score'].value_counts().sort_values(ascending=False)

most_common_score = review_scores.idxmax()

sns.set(style="darkgrid")

plt.figure(figsize=(10, 5))
sns.barplot(x=review_scores.index,
            y=review_scores.values,
            order=review_scores.index,
            palette=["#068DA9" if score == most_common_score else "#D3D3D3" for score in review_scores.index]
            )

plt.title("Penilaian Pelanggan terhadap Pelayanan E-Commerce", fontsize=15)
plt.xlabel("Penilaian/Rating")
plt.ylabel("Jumlah")
plt.xticks(fontsize=12)
plt.show()
```

##### Berdasarkan penilaian tersebut, diperoleh informasi bahwa secara keseluruhan pelanggan memberikan penilaian yang sangat baik terhadap pelayanan yang diberikan oleh e-commerce. Hal tersebut ditunjukkan pada penilaian rating 5 yang mencapai lebih dari 20000 penilaian. Hal tersebut juga signifikn dibandingan dengan rating yang lebih rendah.

### Pertanyaan 5: Bagaimana latar belakang pelanggan yang berbelanja di E-Commerce? Berasal dari kota mana sajakah mereka? Apakah terdapat perbedaan preferensi pembelian antar pelanggan?

Banyak Pelanggan bersasarkan daerah/kota

```python
bystate_df = all_data.groupby(by="customer_state").customer_id.nunique().reset_index()
bystate_df.rename(columns={
    "customer_id": "customer_count"
}, inplace=True)
bystate_df.head()
```

```python
plt.figure(figsize=(12, 6))

most_common_state = bystate_df.loc[bystate_df['customer_count'].idxmax(), 'customer_state']

bystate_df = bystate_df.sort_values(by='customer_count', ascending=False)

sns.barplot(x='customer_state',
            y='customer_count',
            data=bystate_df,
            palette=["#068DA9" if state == most_common_state else "#D3D3D3" for state in bystate_df['customer_state']]
            )

plt.title("Banyak Pelanggan berdasarkan Daerah", fontsize=15)
plt.xlabel("State")
plt.ylabel("Number Customers")
plt.xticks(fontsize=10)
plt.show()
```

##### Perhatikan bahwa pelanggan terbanyak berasal dari daerah dengan kode SP (Sao Paulo).

```python
bycity_df = all_data['customer_city'].value_counts().head(10)

plt.figure(figsize=(12, 6))

most_common_city = bycity_df.idxmax()

bycity_df = bycity_df.sort_values(ascending=False)

sns.barplot(x=bycity_df.index,
            y=bycity_df.values,
            palette=["#068DA9" if city == most_common_city else "#D3D3D3" for city in bycity_df.index]
            )

plt.title("Banyaknya Pelanggan Berdasarkan Kota", fontsize=15)
plt.xlabel("City")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45, fontsize=10)
plt.show()
```

### Pertanyaan 6: Daerah mana sajakah dengan jumlah customer/pelanggan e-commerce terbanyak?

```python
other_state_geolocation = geolocation_df.groupby(['geolocation_zip_code_prefix'])['geolocation_state'].nunique().reset_index(name='count')
other_state_geolocation[other_state_geolocation['count']>= 2].shape
max_state = geolocation_df.groupby(['geolocation_zip_code_prefix','geolocation_state']).size().reset_index(name='count').drop_duplicates(subset = 'geolocation_zip_code_prefix').drop('count',axis=1)
```
